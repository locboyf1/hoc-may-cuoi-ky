```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_path = 'Reduced MNIST Data/Reduced Trainging data'
test_path = 'Reduced MNIST Data/Reduced Testing data'
```

```python
def get_dataframe(path):
    dataframe = []
    for root, dirs, files in os.walk(path):
        if root == train_path:
            continue
        for file in files:
            label = root.split('\\')[1]
            dataframe.append({'image': file, 'label' : label})
        
    return pd.DataFrame(dataframe)

```

```python
df_train = get_dataframe(train_path)
df_test = get_dataframe(test_path)
```
